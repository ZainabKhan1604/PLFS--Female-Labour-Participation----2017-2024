```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
years = sorted(flfp_state["Year_num"].unique())

for y in years:
    # Filter and sort data for the selected year
    data = flfp_state[flfp_state["Year_num"] == y].sort_values("FLFP_rate_15_59", ascending=False)
    
    # Plot bar chart
    plt.figure(figsize=(14,6))
    plt.bar(data["State_Name"], data["FLFP_rate_15_59"], color="seagreen", alpha=0.8)
    plt.xticks(rotation=90, fontsize=9)
    plt.yticks(fontsize=9)
    plt.title(f"Female Labour Force Participation Rate (15–59) by State - {y}", fontsize=12, weight='bold')
    plt.xlabel("State/UT", fontsize=10)
    plt.ylabel("FLFP (%)", fontsize=10)
    plt.tight_layout()

    # ✅ Make sure this line has the SAME indentation level as plt.show() would
    plt.savefig(f"{save_path}\\FLFP_Statewise_{y}.png", dpi=300)
    plt.close()
    plt.show()
```

```python
fwpr_pivot = fwpr_states.pivot_table(
    index="State_Name",
    columns="Year_num",
    values="FWPR_15_59"
).sort_index()
```

```python
fwpr_pivot.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(
    fwpr_pivot,
    cmap="YlGnBu",
    linewidths=0.5,
    linecolor='gray',
    annot=True,
    fmt=".1f",
    cbar_kws={'label': 'FWPR (%)'}
)
plt.title("Female Worker Population Ratio (15–59) by State and Year", fontsize=13, weight='bold')
plt.xlabel("Year")
plt.ylabel("State/UT")
plt.tight_layout()
plt.savefig("FWPR_states.png", dpi=300)
plt.show()
```

```python
literate_lf_india = pd.read_csv(r"C:\Users\ZAINAB KHAN\Desktop\PLFS Project\Results\Literate_Female_Share_LF_India_2017_2024.csv")
literate_notlf_india = pd.read_csv(r"C:\Users\ZAINAB KHAN\Desktop\PLFS Project\Results\Literate_Female_NotLF_India_2017_2024.csv")
```

```python
literate_compare = pd.merge(
    literate_lf_india,
    literate_notlf_india,
    on="Year_num",
    how="inner"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(
    literate_compare["Year_num"],
    literate_compare["Literate_Female_Share_in_LF_India_15_59"],
    marker='o', color='teal', linewidth=2,
    label='Literate Females in Labour Force'
)

plt.plot(
    literate_compare["Year_num"],
    literate_compare["Literate_Female_Share_Not_in_LF_India_15_59"],
    marker='o', color='goldenrod', linewidth=2,
    label='Literate Females not in Labour Force'
)

plt.title("Literate Working-Age Women in India: In vs. Out of Labour Force (2017–2024)", fontsize=12, weight='bold')
plt.xlabel("Year", fontsize=10)
plt.ylabel("Percentage of Literate Females (15–59)", fontsize=10)
plt.xticks(sorted(literate_compare["Year_num"].unique()))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("literate_women.png", dpi=300)
plt.show()
```
