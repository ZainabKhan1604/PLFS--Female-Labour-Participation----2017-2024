```python
import pandas as pd
import glob
import os
```

```python
path = r"C:\Users\ZAINAB KHAN\Desktop\PLFS Project\Cleaned_Data_female"
# Get all CSV files from that folder
all_files = glob.glob(os.path.join(path, "*.csv"))

# Check file list (optional)
print("Files found:", all_files)
```

```python
# Read and concatenate all files
df_all = pd.concat(
    (pd.read_csv(f, low_memory=False) for f in all_files),
    ignore_index=True
)

# Check shape and years included
print(df_all.shape)
print(df_all["Year"].unique())

```

```python
print(df_all["State/UT Code"].nunique())  # should be 34
print(df_all["Year"].value_counts().sort_index())  # see how many records per year
print(df_all.head(1))
```

```python
save_path = r"C:\Users\ZAINAB KHAN\Desktop\PLFS Project\PLFS_Master_2017_2024.csv"
df_all.to_csv(save_path, index=False)
print("✅ Master dataset saved successfully!")
```
